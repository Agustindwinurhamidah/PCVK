```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import math
import os
import glob
```

```python
image  = cv.imread('/content/drive/MyDrive/Images/Images/lena.jpg')
height, width, depth = np.shape(image)
names = np.arange(256)

b = [0]*256
g = [0]*256
r = [0]*256

for y in range (0,height) :
  for x in range (0,width) :
    b[image[y][x][0]] += 1
    g[image[y][x][1]] += 1
    r[image[y][x][2]] += 1

names = np.arange(256)
fig, axs = plt.subplots(1, 3, figsize=[20,5], sharex=True, sharey=True)
fig.suptitle('Histogram RGB plot')
fig.text(0.09, 0.5, 'Jumlah Kemunculan', va='center', rotation='vertical')
fig.text(0.5, 0.04, 'Intensitas Warna', ha='center')
axs[0].bar(names, r, color = 'red')
axs[1].bar(names, g, color = 'green')
axs[2].bar(names, b, color = 'blue')
```

```python
from numpy.lib.histograms import histogram

lena_original = cv.imread('/content/drive/MyDrive/Images/Images/lena.jpg')
colors = ("red", "green", "blue")
channel_ids = (2, 1, 0)

plt.xlim([0,256])
for channel_id, c in zip(channel_ids, colors):
  histogram, bin_edges = np.histogram(
      lena_original[:, :, channel_id], bins=256, range = (0,256)
  )
  plt.plot(bin_edges[0:-1], histogram, color=c)

  plt.xlabel("Intensitas Warna")
  plt.xlabel("Jumlah Kemunculan")

  plt.show()
  plt.show()
```
